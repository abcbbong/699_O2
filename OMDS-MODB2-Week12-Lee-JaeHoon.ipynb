# 2. Weekly graph question

Suppose you plot the number visitors to a museum over the seven days of a week; then again the following week.  Can you suggest the advantages and disadvantages of the following two ways of presenting the data?

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
time_series_0 = np.array([100, 110, 130, 120, 100, 90, 80])
time_series_1 = np.array([70, 90, 100, 100, 120, 110, 90])
fig, axs = plt.subplots(1, 2)
axs[0].plot(time_series_0)
axs[1].plot(time_series_1)
plt.subplots_adjust(wspace = 0)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

```python
plt.plot(time_series_0)
plt.plot(time_series_1)
plt.xlabel("Day of the week")
plt.ylabel("Number of visitors")
```

### Answer

- **Method 1: Continuous Line Graph** – Plot both weeks as a single continuous line (x-axis: Day 1 to Day 14; y-axis: Number of visitors).
- **Method 2: Overlapping Bar Chart by Day of the Week** – Group by days (Monday to Sunday) and overlay bars for both weeks (different colors for each week).

#### Advantages and Disadvantages

| Method                          | Advantages                                                                 | Disadvantages                                                                 |
|---------------------------------|-----------------------------------------------------------------------------|-------------------------------------------------------------------------------|
| **Method 1: Continuous Line Graph** | - Shows overall trends across the entire period (e.g., rising or falling visitor numbers over 14 days).<br>- Emphasizes temporal flow and continuity (e.g., easy to see if patterns carry over from one week to the next).<br>- Simple and space-efficient for displaying sequential data. | - Harder to compare specific days across weeks (e.g., comparing Mondays requires mental alignment).<br>- May obscure weekly cycles (e.g., weekend peaks blend into the line without clear separation).<br>- Risk of visual clutter if the line becomes too long or noisy. |
| **Method 2: Overlapping Bar Chart by Day** | - Enables direct day-to-day comparisons between weeks (e.g., side-by-side bars for each Monday).<br>- Highlights weekly patterns and differences (e.g., emphasizing higher weekend visits and variations).<br>- More intuitive for categorical analysis, with colors aiding quick insights. | - Loses the sense of continuous time (e.g., no view of how the end of Week 1 flows into Week 2).<br>- Can appear cluttered with overlapping bars if differences are subtle.<br>- Less effective for showing long-term trends beyond weekly cycles. |

#### Recommendation
If the goal is to compare weekly patterns (e.g., day-specific changes), Method 2 is preferable for clarity and emphasis on differences. For overall trends over time, Method 1 works better. Choose based on the story you want to tell with the data—focus on comparability and avoid distortion.

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

# --- 1. Data Preparation ---
# Reconstruct the data based on the visual chart.
# Using np.nan for missing data points is crucial for correctly plotting discontinuous lines.
years = np.array([2008, 2009, 2010, 2011, 2012, 2013, 2014])
data = {
    'A': np.array([380, 420, 450, 425, 400, 350, 300]),
    'B': np.array([350, 400, 425, 380, 350, 300, 250]),
    'C': np.array([np.nan, np.nan, 120, 170, 200, 260, 170]),
    'D': np.array([np.nan, np.nan, np.nan, 175, 280, 220, 215]),
    'E': np.array([np.nan, np.nan, np.nan, np.nan, np.nan, 120, 210])
}

# --- 2. Figure and Axes Setup ---
# Create a figure and a set of subplots. Using an object-oriented approach (fig, ax)
# allows for greater control over the plot elements.
fig, ax = plt.subplots(figsize=(10, 6))

# --- 3. Plotting the Data Series ---
# Iterate through the data dictionary to plot each product line.
# Using a neutral grey color for all lines directs the viewer's attention
# to the highlighted elements (recommended range and AVG point).
for label, prices in data.items():
    ax.plot(years, prices, color='grey', linewidth=2.5, solid_capstyle='round')

# --- 4. Adding Contextual Layers ---
# Add the shaded rectangle for the "Recommended range".
# 'zorder=0' places the rectangle behind all other plot elements.
ax.axhspan(150, 200, color='lightsteelblue', alpha=0.4, zorder=0)

# Plot the specific average price point.
# Using a distinct color and marker makes it a focal point.
ax.plot(2014, 223, 'o', color='#004488', markersize=8, label='AVG')

# --- 5. Customizing Axes and Spines ---
# Set the limits for the y-axis and x-axis.
ax.set_ylim(0, 500)
# Add some padding to the x-axis limits for better spacing.
ax.set_xlim(2007.5, 2014.5)

# Remove the top and right spines for a cleaner, modern look.
# This follows the principle of maximizing the data-ink ratio.
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('grey')
ax.spines['bottom'].set_color('grey')

# Customize tick parameters for a polished appearance.
ax.tick_params(axis='x', colors='grey', length=6)
ax.tick_params(axis='y', colors='grey', length=6)

# Format y-axis labels to appear as currency ($).
formatter = mticker.FuncFormatter(lambda x, pos: f'${int(x)}')
ax.yaxis.set_major_formatter(formatter)

# Set custom x-axis ticks to match the original chart.
ax.set_xticks(years)

# --- 6. Adding Annotations and Labels ---
# Add labels for each line series directly on the plot.
# This technique, called direct labeling, is often clearer than a separate legend.
ax.text(2008, 400, 'A', fontsize=12, color='dimgray', ha='center')
ax.text(2008, 330, 'B', fontsize=12, color='dimgray', ha='center')
ax.text(2010.2, 100, 'C', fontsize=12, color='dimgray', ha='center')
ax.text(2011, 155, 'D', fontsize=12, color='dimgray', ha='center')
ax.text(2013, 100, 'E', fontsize=12, color='dimgray', ha='center')

# Add text for the "Recommended range" and "AVG" point.
ax.text(2008, 160, 'Recommended range', fontsize=12, color='darkslateblue', ha='left')
ax.text(2013.6, 225, 'AVG', fontsize=11, color='#004488', ha='right')

# Add the subtitle inside the plot area.
ax.set_title('Retail price over time', loc='left', fontsize=14, color='dimgray', pad=20)
ax.set_xlabel('Year', fontsize=12, color='grey', labelpad=10)
ax.set_ylabel('Average price', fontsize=12, color='grey', labelpad=10)

# --- 7. Crafting the Main Title ---
# Use fig.text() for positioning text relative to the entire figure.
# Adjusted x positions based on calculated text widths to prevent overlapping.
fig.text(0.125, 0.96, 'To be competitive, we recommend introducing our product',
         fontsize=16, color='darkslateblue', ha='left', va='top')
fig.text(0.779, 0.96, ' below',
         fontsize=16, color='darkslateblue', ha='left', va='top', style='italic')
fig.text(0.125, 0.92, 'the $223 average price point in the',
         fontsize=16, color='darkslateblue', ha='left', va='top', style='italic')
fig.text(0.520, 0.92, ' $150–$200 range',
         fontsize=16, color='darkslateblue', ha='left', va='top', weight='bold')

# --- 8. Highlighting a Specific Y-axis Tick ---
# Get the current y-axis ticks and add 150 to the list.
yticks = list(ax.get_yticks())
if 150 not in yticks:
    yticks.append(150)
yticks.sort()
ax.set_yticks(yticks)

# Redraw the plot to apply the new ticks before getting the labels.
fig.canvas.draw()

# Iterate through the new y-tick labels and color the '$150' label.
for label in ax.get_yticklabels():
    if label.get_text() == '$150':
        label.set_color('darkslateblue')
        label.set_weight('bold')

# --- 9. Final Adjustments and Display ---
# Adjust layout to prevent labels from overlapping.
# The 'rect' parameter creates padding at the top of the figure for the main title.
plt.tight_layout(rect=[0, 0, 1, 0.9])

# Display the plot.
plt.show()
```
